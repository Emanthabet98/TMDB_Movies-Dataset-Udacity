Noticed from the previous cell that the 'runtime', 'budget', 'revenue', 'budget_adj', and 'revenue_adj' have zero values, which is'nt convenient

```python
def row_zero(df, col_name):
    """
    check the rows with zero value.
    
    Arges:
        (pandas datafrma) df- the data that needed to check the zero values .
        (str) col_name- the name of the column needed to check the zero values.
    Returns:
        (array) index- array contines the index of the zero values. 
        (tuple) shape- contines the number of the the zero values.
        
    """
    zero = df[col_name] == 0
    index = df[zero].index.values
    shape = df[zero].shape
    return index, shape
```

## Q5: what's the Top and least 10 movies based on the revenue?

```python
def top_10_movies(df, col_name):
    """
    Sort the movies titles and return the the top 10 movies title dapanding on a certain column.
    
    Return:
        x_axis- dataframe contains the top 10 movies titles.
        y_axis- dataframe contains the top 10 values of a certain column which related to the movies titles.
    """
    movies_and_col = df[["original_title", col_name]]
    
    x_axis = movies_and_col.sort_values(by = col_name, ascending=False).head(10).original_title
    y_axis = movies_and_col.sort_values(by = col_name, ascending=False).head(10)[col_name]
    
    return x_axis, y_axis
```

```python
def least_10_movies(df, col_name):
    """
    Sort the movies titles and return the the least 10 movies title dapanding on a certain column.
    
    Return:
        x_axis- dataframe contains the least 10 movies titles.
        y_axis- dataframe contains the least 10 values of a certain column which related to the movies titles.
    """
    movies_and_col = df[["original_title", col_name]]
    
    x_axis = movies_and_col.sort_values(by = col_name, ascending=False).tail(10).original_title
    y_axis = movies_and_col.sort_values(by = col_name, ascending=False).tail(10)[col_name]
    
    return x_axis, y_axis
```

```python
#Apply top_10_movies(df, col_name) function on 'revenue' column.
x_axis = top_10_movies(movies_df, 'revenue')[0]
y_axis = top_10_movies(movies_df, 'revenue')[1]

plt.figure(figsize=(20, 10))
plt.barh(x_axis, y_axis)

plt.title('Top 10 Movies Based On Revenue', fontsize=25)

plt.ylabel('Movies\' Title ', fontsize=25)
plt.xlabel('Total Revenue', fontsize=25)

plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.show()

```

```python
#Apply least_10_movies(df, col_name) function on 'revenue' column.

x_axis = least_10_movies(movies_df, 'revenue')[0]
y_axis = least_10_movies(movies_df, 'revenue')[1]

plt.figure(figsize=(20, 10))
plt.barh(x_axis, y_axis)

plt.title('Least 10 Movies Based On Revenue', fontsize=25)

plt.ylabel('Movies\' Title ', fontsize=25)
plt.xlabel('Total Revenue', fontsize=25)

plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.show()
```
